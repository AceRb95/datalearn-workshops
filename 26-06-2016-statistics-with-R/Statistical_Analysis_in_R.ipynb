## Shape

### Skewness

A distribution is said to be right-skewed (or positively skewed) if the right tail seems to be stretched from the center. A left-skewed (or negatively skewed) distribution is stretched to the left side.

```r
require(e1071)
```

```r
skewness(diamonds$x)
```

### Kurtosis

It is a measure of peakedness and tailedness of the distribution.

Some distributions tend to have a flat shape with thin tails. These are called platykurtic, and an example of a platykurtic distribution (kurtosis < 0) is the uniform distribution; see Section 6.2. On the other end of the spectrum are distributions with a steep peak, or spike, accompanied by heavy tails; these are called leptokurtic (kurtosis > 0). Examples of leptokurtic distributions are the Laplace distribution and the logistic distribution. See Section 6.5. In between are distributions (called mesokurtic) with a rounded peak and moderately sized tails

```r
kurtosis(diamonds$x)
```

As we know, random numbers are described by a distribution. That is, some function which specifies the probability
that a random number is in some range. For example P(a < X ≤ b). Often this is given by a probability density (in the continuous case) or by a function P(X = k) = f(k) in the discrete case

R has different functions to generate density, distribution, quantile and random deviates for a distribution.

|name     |         description             |
|---------|---------------------------------|

|pname( ) |	cumulative density function     |
|qname( ) |	quantile function               |
|rname( ) |	random deviates                 |

clear difference between density (dname for continuous), probability (dname for discrete) and cumulative probability (pname)

## difference between density (dnorm) and cumulative probability (pnorm)

The density function is (for a continuous random variable) a function that describes the relative probability of the values taken by the variable (it is not probability! the word relative is important).

So where the density is high, values are relatively more probable, and where it's low, values are relatively less probable.

Areas under the density, between values on the x-axis, do give probabilities of lying in those intervals

```r
xseq <- seq(-4, 4, .01)
densities <- dnorm(xseq, 0, 1)
cumulative <- pnorm(xseq, 0, 1)
randomdeviates <- rnorm(1000, 0, 1)
 
par(mfrow=c(1,2))

plot(xseq, densities, col="darkgreen",xlab="", ylab="Density", type="l",lwd=2, cex=2, main="PDF of Standard Normal", cex.axis=.8)
abline(v=2)
abline(h=dnorm(2))
text(2, dnorm(2), paste0("2, ", round(dnorm(2),2)), pos=4)
abline(v=-1.6)
abline(h=dnorm(-1.6))
text(-1.6, dnorm(-1.6), paste0("-1.6, ", round(dnorm(-1.6),2)), pos=4)

plot(xseq, cumulative, col="darkorange", xlab="", ylab="Cumulative Probability",
     type="l",lwd=2, cex=2, main="CDF of Standard Normal", cex.axis=.8)

# hist(randomdeviates, main="Random draws from Std Normal", cex.axis=.8, xlim=c(-4,4))
```

The height is the relative probability; so the height at 2 (about 0.05) is saying that a value very close to 2 is approximately half as probable as a value very close to -1.6 (where the density is about 0.11). [The density at 2 is not the probability of getting 2, because density is not itself probability]

We could work out the probability of getting a value between -1.6 and 2 by finding the area of the curve between those values (that area turns out to be 0.922).

# Binomial
They have the distribution of the number
of successes in n independent Bernoulli trials where a Bernoulli trial results in success or failure, success with
probability p.

![binomial distribution](https://wikimedia.org/api/rest_v1/media/math/render/svg/8f1eb36a9ece93abf99484b655139921da8a843b)

When dealing with discrete random variables, such as counts, we are not dealing with density, but an actual probability function. The height of the probability function is actually the probability of the given value.

```r
x=rbinom(100, 10, .5)
hist(x, probability=T)
points(0:10, dbinom(0:10, 10, .5), type="h", lwd=3)
points(0:10, dbinom(0:10, 10, .5), type="p", lwd=3)
```

# Fitting Distributions

Histogram is not a good way, because its visual perception changes with bin size

```r
set.seed(123)
x <- runif(100,  min=0, max=1)
par(mfrow=c(1,3))
hist(x,breaks=seq(0,1,.1),col=gray(.9))
hist(x,breaks=seq(0,1,.2),col=gray(.9))
hist(x,breaks=c(0,.35,.7,1),col=gray(.9))
```

Cumulative Distribution Function (CDF)

```r
set.seed(123)
x <- runif(100,  min=0, max=1)
plot(ecdf(x))
curve(punif(x), from=0, to=1, n=100, add=T, col="red", lwd=2)
```

Q-Q Plot

The q-q plot for uniform data is very similar to the empirical CDF graphic, except with the axes reversed. The q-q plot provides a visual comparison of the sample quantiles to the corresponding theoretical quantiles. In general, if the points in a q-q plot depart from a straight line, then the assumed distribution is called into question.

```r
set.seed(123)
x <- rnorm(100)
qqnorm(x)
qqline(x)
```
